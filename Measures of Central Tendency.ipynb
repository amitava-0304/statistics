<h1> Measures of Central Tendency</h1>
<p>Measures of central tendency are statistical metrics that describe the center or typical value of a dataset. They summarize data with a single value that represents the middle or center of the distribution. The main measures include:</p>

<h2> Mean (Arithmetic Average)</h2>
<p><strong>Definition</strong>: The mean is the sum of all data points divided by the number of data points.</p>



<p><strong>Example</strong>: For the dataset Ages = {1, 3, 4, 5}:</p>
<p><code>x = (1 + 3 + 4 + 5) / 4 = 13 / 4 = 3.25</code></p>

<p><strong>Use Case</strong>: The mean is suitable for data without extreme outliers, as outliers can significantly skew the mean.</p>

<hr>

<h2> Median</h2>
<p><strong>Definition</strong>: The median is the middle value of a dataset when the numbers are arranged in ascending or descending order.</p>

<p><strong>How to Find It</strong>:<br>
If <strong>N</strong> (the number of data points) is odd, the median is the middle value.<br>
If <strong>N</strong> is even, the median is the average of the two middle numbers.</p>

<p><strong>Examples</strong>:</p>
<ul>
    <li>For the dataset Ages = {1, 3, 4, 5, 100} (with an outlier):
        <br>Sorted dataset: {1, 3, 4, 5, 100}
        <br>Median: <strong>4</strong> (middle value in an odd-numbered dataset)</li>
    <li>For the dataset Ages = {4, 3, 1, 5, 100, 200} (even number of elements):
        <br>Sorted dataset: {1, 3, 4, 5, 100, 200}
        <br>Median: <strong>(4 + 5) / 2 = 4.5</strong></li>
</ul>

<p><strong>Use Case</strong>: The median is ideal for skewed distributions or datasets with outliers, as it is less affected by extreme values.</p>

<hr>

<h2> Mode</h2>
<p><strong>Definition</strong>: The mode is the value that appears most frequently in a dataset.</p>

<p><strong>Characteristics</strong>: A dataset can be:<br>
<ul>
    <li><strong>Unimodal</strong> (one mode)</li>
    <li><strong>Bimodal</strong> (two modes)</li>
    <li><strong>Multimodal</strong> (multiple modes)</li>
    <li><strong>No mode</strong> (if no number repeats)</li>
</ul></p>

<p><strong>Examples</strong>:</p>
<ul>
    <li><strong>Dataset 1</strong>: Ages = {1, 3, 4, 4, 5, 5, 5}
        <br>The value <strong>5</strong> appears <strong>3 times</strong>, which is more frequent than any other value in the dataset.
        <br><strong>Mode</strong>: 5
    </li>
    <li><strong>Dataset 2</strong>: Ages = {1, 2, 3, 4, 5}
        <br>Each number appears only once, so there is <strong>no mode</strong>.
    </li>
    <li><strong>Dataset 3</strong>: Ages = {3, 3, 5, 5, 7, 7}
        <br>There are two values (3 and 5) that appear twice each, so this dataset is <strong>bimodal</strong>, with modes <strong>3</strong> and <strong>5</strong>.
    </li>
</ul>

<p><strong>Use Case</strong>: The mode is particularly useful when analyzing <strong>categorical data</strong>, such as the most common shirt sizes sold in a store (e.g., S, M, L, XL).</p>

<hr>

<h3> Choosing the Right Measure</h3>
<ul>
    <li><strong>Mean</strong> is best for symmetric distributions without outliers.</li>
    <li><strong>Median</strong> is preferred for skewed distributions or datasets with outliers.</li>
    <li><strong>Mode</strong> is valuable for categorical data or identifying the most common occurrence in a set.</li>
</ul>

<p>Each measure provides different insights into the data's central value, and the choice depends on the distribution and nature of the data.</p>
