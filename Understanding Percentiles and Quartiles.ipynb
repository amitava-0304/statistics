# 📊 Understanding Percentiles

A **percentile** is a value in a dataset below which a given percentage of observations fall. It is a measure used to understand the relative standing of a data point within a distribution.

## 🚀 Key Points:
- Percentile indicates the position or rank of a value in a dataset relative to others.
- For example, if you are in the **80th percentile**, it means that 80% of the data points are below your value and 20% are above it.

## 🔢 Percentile Calculation:
A percentile rank is calculated as the percentage of values in the dataset that are less than or equal to the given value.

### 🧑‍🏫 Example:
Consider the following dataset of test scores:
60, 70, 75, 80, 85, 90, 95, 100    




<h1>How to Calculate Percentiles: A Step-by-Step Example</h1>

<p>Let's go through a practical example using a dataset of test scores. The goal is to calculate the percentile for a given value, say <strong>95</strong>, in the following dataset:</p>

<h2>Test Scores:</h2>
<pre>[60, 70, 75, 80, 85, 90, 95, 100]</pre>

<h3>Step 1: Identify the Number of Values Below 95</h3>
<p>We need to count how many data points are below the value of <strong>95</strong> in the dataset. From the sorted dataset, the values below 95 are:</p>

<pre>[60, 70, 75, 80, 85, 90]</pre>

<p>There are <strong>6 values</strong> below 95.</p>

<h3>Step 2: Total Number of Data Points</h3>
<p>The total number of data points in this dataset is <strong>8</strong>.</p>

<h3>Step 3: Apply the Percentile Formula</h3>
<p>Now, we apply the formula for calculating the percentile:</p>

<pre>Percentile of 95 = (Number of values below 95 × 100) / Total number of data points</pre>

<p>Substitute the values:</p>

<pre>Percentile of 95 = (6 × 100) / 8 = 600 / 8 = 75</pre>

<h3>Interpretation:</h3>
<p>The percentile corresponding to the value 95 is <strong>75th percentile</strong>. This means that 95 is greater than or equal to <strong>75%</strong> of the values in the dataset. In other words, <strong>75%</strong> of the students scored below 95 on the test.</p>

<h2>📈 Conclusion:</h2>
<ul>
    <li>Percentiles are useful for understanding the relative position of a value in a dataset.</li>
    <li>In this case, the <strong>75th percentile</strong> shows that 75% of the students scored less than or equal to 95.</li>
    <li>This calculation can help assess how a given score compares to the rest of the data.</li>
</ul>

<h3>💡 Key Takeaways:</h3>
<ul>
    <li>Percentiles allow us to measure the distribution of data points.</li>
    <li>In the 75th percentile example, 95 is better than or equal to 75% of the scores in the dataset.</li>
</ul>

# 🔢 Manual Method (Interpolation):

<h1>Calculating the 75th Percentile</h1>

<p>To calculate the 75th percentile of the given dataset:</p>

<pre>[60, 70, 75, 80, 85, 90, 95, 100]</pre>

<h2>Steps to Calculate the 75th Percentile:</h2>

<h3>1. Sort the Data (if not already sorted):</h3>
<p>The dataset is already sorted:</p>
<pre>[60, 70, 75, 80, 85, 90, 95, 100]</pre>

<h3>2. Determine the Position of the 75th Percentile:</h3>
<p>To find the 75th percentile, we use the following formula to calculate the position:</p>

<pre>𝑃ₖ = (𝑘 / 100) × (𝑁 + 1)</pre>

<p>Where:</p>
<ul>
    <li><strong>𝑃ₖ</strong> is the position of the <strong>k-th percentile</strong>.</li>
    <li><strong>𝑘 = 75</strong> (since we want the 75th percentile).</li>
    <li><strong>𝑁 = 8</strong> (total number of data points).</li>
</ul>

<p>Now, substituting the values:</p>

<pre>𝑃₇₅ = (75 / 100) × (8 + 1) = 0.75 × 9 = 6.75</pre>

<p>This means that the 75th percentile lies between the <strong>6th</strong> and <strong>7th</strong> data points.</p>

<h3>3. Interpolate Between the 6th and 7th Data Points:</h3>
<p>The 6th data point is 90, and the 7th data point is 95. Since the position is 6.75, we will interpolate between these two values.</p>

<h4>Interpolation Formula:</h4>
<pre>Percentile Value = Lower Value + (Fraction × (Upper Value - Lower Value))</pre>

<p>Substituting the values:</p>

<pre>Percentile Value = 90 + (0.75 × (95 - 90))</pre>
<pre>Percentile Value = 90 + (0.75 × 5) = 90 + 3.75 = 93.75</pre>

<h3>Result:</h3>
<p>The <strong>75th percentile</strong> of the dataset is <strong>93.75</strong>.</p>

<p>This means that <strong>93.75</strong> is the value below which 75% of the data points fall.</p>

<h2>📊 Conclusion:</h2>
<ul>
    <li>The 75th percentile value is <strong>93.75</strong>.</li>
    <li>This means 75% of the values in the dataset are less than or equal to 93.75.</li>
</ul>

<h3>🔑 Key Insights:</h3>
<ul>
    <li>The 75th percentile shows the point below which three-quarters of the data lies.</li>
    <li>By using interpolation, we obtained the exact value (93.75) between the 6th and 7th data points.</li>
</ul>

<h1>📊 The Difference Between Manual and NumPy Percentile Calculation</h1>

<div>
  <h2>🔢 Manual Method (Interpolation):</h2>
  <p>When you calculate the 75th percentile manually, you're <strong>interpolating</strong> between the 6th and 7th values in the sorted dataset. This interpolation assumes that the 75th percentile lies between two data points, so you calculate a fractional value (93.75 in this case).</p>
</div>

<div>
  <h2>💻 NumPy's Method:</h2>
  <p>NumPy's <code>np.percentile</code> function uses a different method to calculate percentiles, called <strong>linear interpolation</strong> by default, but it computes percentiles in a way that might not always interpolate between data points in the same manner as manual methods. Specifically, NumPy uses a method known as <strong>"nearest rank"</strong> or <strong>"linear interpolation"</strong>, depending on the parameter settings, and the way it rounds or selects data points may lead to a value that is directly one of the data points (in this case, 95).</p>
</div>

<div>
  <h2>🤔 Why 95.0 Instead of 93.75?</h2>
  <ul>
    <li><strong>📊 Rank-based Method:</strong> The percentile is calculated based on <strong>ranks</strong>, and in some cases (like this one), it will return the value at the <strong>nearest rank</strong> that corresponds to the percentile. Since the position 6.75 is closer to the 7th data point (95) than to the 6th (90), NumPy will return 95.</li>
    <li><strong>🔄 Interpolation Logic in NumPy:</strong> The exact interpolation method used by <code>np.percentile</code> for the "75th percentile" might not always produce a <strong>fractional result</strong> but instead may <strong>round</strong> or adjust based on how the function calculates the position.</li>
  </ul>
</div>

<div>
  <h2>🔑 Key Takeaways:</h2>
  <ul>
    <li><strong>Manual calculation</strong> gives you a fractional percentile (93.75) by <strong>interpolating</strong> between two points.</li>
    <li><strong>NumPy's percentile function</strong> might <strong>round</strong> the result or choose a <strong>nearest rank</strong> (95 in this case).</li>
    <li>Different methods of percentile calculation are valid, but they <strong>reflect different approaches</strong> (interpolation vs rank-based).</li>
  </ul>
</div>


## 📐 Quartiles Definition

**Quartiles** are values that divide a dataset into four equal parts. The data is divided such that each part contains the same number of observations. Quartiles help in understanding the distribution of the data.

### Types of Quartiles:
1. **Q1 (First Quartile):** The value that separates the lowest 25% of the data from the rest. It is the **25th percentile**.
2. **Q2 (Second Quartile or Median):** The middle value, dividing the data into two halves. It is the **50th percentile**.
3. **Q3 (Third Quartile):** The value that separates the lowest 75% of the data from the top 25%. It is the **75th percentile**.

These quartiles help in understanding the spread and distribution of the data.


## ✍️ Quartile Formula

The generalized formula for finding a **quartile (Q_r)** in a grouped dataset is:

$$
Q_r = l_1 + \left( \frac{i \cdot n}{4} - c_f \right) \cdot \frac{(l_2 - l_1)}{f}
$$

Where:
- \( Q_r \) = \( r^{th} \) quartile.
- \( l_1 \) and \( l_2 \) = Lower and upper limits of the class containing the quartile.
- \( f \) = Frequency of the class containing the quartile.
- \( c_f \) = Cumulative frequency of classes before the quartile.
- \( n \) = Total number of data points.
- \( i \) = Index for the quartile (1 for Q1, 2 for Q2, etc.)

### **Example:**
For example, to calculate **Q1**, use the formula:

$$
Q1 = l_1 + \left( \frac{n}{4} - c_f \right) \cdot \frac{(l_2 - l_1)}{f}
$$

You can apply similar formulas for **Q2** and **Q3**.


## 📈 Visualizing Quartiles with a Bell Curve

Let's now visualize the data using a **bell curve** with the **Iris dataset**'s **sepal_length** and mark the **quartiles (Q1, Q2, Q3)** on the curve.

We will use the following steps:
1. **Plot the histogram** of the data.
2. **Overlay the KDE (Kernel Density Estimate)**.
3. **Mark the quartiles** (Q1, Q2, Q3) using vertical lines.
4. **Annotate the quartile values** on the graph.

### Code for Bell Curve with Quartiles:





````python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Load Iris dataset
iris = sns.load_dataset('iris')

# Select the 'sepal_length' column
data = iris['sepal_length']

# Calculate Quantiles (Q1, Q2, Q3)
quantiles = np.percentile(data, [25, 50, 75])
q1, q2, q3 = quantiles

# Create Bell Curve Plot
plt.figure(figsize=(10, 6))

# Plot histogram and KDE
sns.histplot(data, kde=True, bins=20, color='blue', alpha=0.6, label='Data Distribution')

# Add vertical lines for Q1, Q2, Q3
plt.axvline(q1, color='red', linestyle='--', label=f'Q1 (25th): {q1:.2f}')
plt.axvline(q2, color='green', linestyle='--', label=f'Q2 (50th, Median): {q2:.2f}')
plt.axvline(q3, color='orange', linestyle='--', label=f'Q3 (75th): {q3:.2f}')

# Annotate quantiles
plt.text(q1, 5, f'Q1 ({q1:.2f})', horizontalalignment='center', color='red', fontsize=12)
plt.text(q2, 5, f'Q2 ({q2:.2f})', horizontalalignment='center', color='green', fontsize=12)
plt.text(q3, 5, f'Q3 ({q3:.2f})', horizontalalignment='center', color='orange', fontsize=12)

# Labels and title
plt.title('Bell Curve with Quantiles for Sepal Length', fontsize=16)
plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

# Show the plot
plt.show()

# Print Quantile Values
print(f"Q1: {q1:.2f}, Q2: {q2:.2f}, Q3: {q3:.2f}")
````

- Vertical lines marking **Q1**, **Q2**, and **Q3**.
- Annotations with the **quantile values**.
